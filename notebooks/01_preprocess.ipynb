```python
from PIL import Image, ImageOps
import os

# Correct paths for your project
INPUT_FOLDER = r"D:\DataScience\data\raw_images"
OUTPUT_FOLDER = r"D:\DataScience\data\processed_images"

SIZE = (224, 224)

def augment_and_save(img_path, save_folder, base_name):
    img = Image.open(img_path).convert("RGB").resize(SIZE)

    variants = {
        "original": img,
        "rotate90": img.rotate(90, expand=True),
        "rotate180": img.rotate(180, expand=True),
        "flipped": ImageOps.mirror(img),
    }

    for tag, variant in variants.items():
        save_path = os.path.join(save_folder, f"{base_name}_{tag}.jpg")
        variant.save(save_path, format="JPEG", quality=85)

# Process all products
for product_name in os.listdir(INPUT_FOLDER):
    input_product_folder = os.path.join(INPUT_FOLDER, product_name)

    if not os.path.isdir(input_product_folder):
        continue

    output_product_folder = os.path.join(OUTPUT_FOLDER, product_name)
    os.makedirs(output_product_folder, exist_ok=True)

    for img_file in os.listdir(input_product_folder):
        if img_file.lower().endswith((".jpg", ".jpeg", ".png")):
            img_path = os.path.join(input_product_folder, img_file)
            base_name = os.path.splitext(img_file)[0]
            augment_and_save(img_path, output_product_folder, base_name)

    print(f"✅ Done: {product_name}")

print("\n🎉 All images preprocessed!")
```
